```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
```

Load prepared dataset

```python
prepared_data = pd.read_excel(r'dataset/prepared_dataset/BRICS_all_areas_prepared_dataset.xlsx')
prepared_data = prepared_data.iloc[:, 1:]
prepared_data
```

Delete duplicate columns

```python
data_without_duplicate = prepared_data.drop(labels=[
    "Survey mean consumption or income per capita, total population (2011 PPP $ per day).1",
    "Survey mean consumption or income per capita, bottom 40% of population (2011 PPP $ per day).1",
    'Proportion of people living below 50 percent of median income (%).1',
    'Poverty headcount ratio at national poverty lines (% of population).1',
    'Poverty headcount ratio at $5.50 a day (2011 PPP) (% of population).1',
    'Poverty headcount ratio at $3.20 a day (2011 PPP) (% of population).1',
    'Poverty headcount ratio at $1.90 a day (2011 PPP) (% of population).1',
    'Poverty gap at $5.50 a day (2011 PPP) (%).1',
    'Poverty gap at $3.20 a day (2011 PPP) (%).1',
    'Poverty gap at $1.90 a day (2011 PPP) (%).1',
    'Multidimensional poverty intensity (average share of deprivations experienced by the poor).1',
    'Multidimensional poverty index (scale 0-1).1',
    'Multidimensional poverty headcount ratio (% of total population).1',
    'Income share held by third 20%.1',
    'Income share held by second 20%.1',
    'Income share held by lowest 20%.1',
    'Income share held by lowest 10%.1',
    'Income share held by highest 20%.1',
    'Income share held by highest 10%.1',
    'Income share held by fourth 20%.1',
    'Gini index (World Bank estimate).1',
    "Annualized average growth rate in per capita real survey mean consumption or income, total population (%).1",
    "Annualized average growth rate in per capita real survey mean consumption or income, bottom 40% of population (%).1"
], axis=1)
data_without_duplicate
```

Check data types

```python
data_without_duplicate.dtypes
```

```python
data_without_duplicate.describe()
```

Filling in empty values

```python
data_filled = data_without_duplicate.copy()
data_filled = data_filled.bfill()
data_filled = data_filled.ffill()
data_filled
```

Drop useless columns

```python
columns_to_drop = data_filled.columns[data_filled.nunique() == 1]
data_without_useless_columns = data_filled.drop(columns=columns_to_drop)
data_without_useless_columns
```

Create corr matrix

```python
corr_matrix = data_without_useless_columns.corr()
corr_matrix
```

View target feature correlation

```python
corr_matrix["GDP per capita (current US$)"].sort_values(ascending=False)
```

Rescale data

```python
scaler = MinMaxScaler(feature_range=(0, 1))
rescaled_data = pd.DataFrame(scaler.fit_transform(data_without_useless_columns))
rescaled_data = rescaled_data.set_axis([column for column in data_without_useless_columns], axis=1)
rescaled_data = rescaled_data.iloc[1:, :].reset_index().iloc[:, 1:]
rescaled_data
```

Save the resulting dataset

```python
rescaled_data.to_excel(r'dataset/cleared_dataset/BRICS_all_areas_cleared_dataset.xlsx')
```
