File for creating new datasets based on the prepared dataset.
Creating datasets with different methods for filling empty fields, content, etc.

```python
import pandas as pd
from IPython.display import display
from sklearn.preprocessing import MinMaxScaler
```

```python
all_datasets = {}
```

Load prepared dataset

```python
prepared_data = pd.read_excel(r'dataset/prepared_dataset/main_prepared_dataset.xlsx')
prepared_data = prepared_data.iloc[:, 1:]
prepared_data
```

Check data types

```python
prepared_data.dtypes
```

```python
prepared_data.describe()
```

Formation of datasets based on different types of filling empty cells

```python
classic_filling = prepared_data.copy()
classic_filling = classic_filling.bfill()
classic_filling = classic_filling.ffill()
classic_filling
```

Collecting signs - constants

```python
columns_to_drop = classic_filling.columns[classic_filling.nunique() == 1]
data_without_useless_columns = classic_filling.drop(columns=columns_to_drop)
all_datasets['bfill_ffill'] = data_without_useless_columns
data_without_useless_columns
```

We remove the characteristics of the constant from the main dataset
(since during interpolation these values may become non-constant)

```python
unfilled_data = prepared_data.drop(columns=columns_to_drop)
unfilled_data
```

Linear interpolation

```python
linear_interpolation_dataset = unfilled_data.copy()
linear_interpolation_dataset = linear_interpolation_dataset.interpolate(method='linear')
linear_interpolation_dataset = linear_interpolation_dataset.bfill()
linear_interpolation_dataset = linear_interpolation_dataset.ffill()
all_datasets['linear_interpolation'] = linear_interpolation_dataset
linear_interpolation_dataset
```

Quadratic interpolation

```python
quadratic_interpolation_dataset = unfilled_data.copy()
for i in quadratic_interpolation_dataset:
    try:
        quadratic_interpolation_dataset[i].interpolate(method='quadratic', inplace=True)
    except ValueError as e:
        print(i)

quadratic_interpolation_dataset = quadratic_interpolation_dataset.bfill()
quadratic_interpolation_dataset = quadratic_interpolation_dataset.ffill()
all_datasets['quadratic_interpolation'] = quadratic_interpolation_dataset
quadratic_interpolation_dataset
```

Cubic interpolation

```python
cubic_interpolation_dataset = unfilled_data.copy()
for i in cubic_interpolation_dataset:
    try:
        cubic_interpolation_dataset[i].interpolate(method='cubic', inplace=True)
    except ValueError as e:
        print(i)

cubic_interpolation_dataset = cubic_interpolation_dataset.bfill()
cubic_interpolation_dataset = cubic_interpolation_dataset.ffill()
all_datasets['cubic_interpolation'] = cubic_interpolation_dataset
cubic_interpolation_dataset
```

Polynomial interpolation (odd only)

```python
for degree in range(5, 12, 2):
    polynomial_interpolation_dataset = unfilled_data.copy()
    for i in polynomial_interpolation_dataset:
        try:
            polynomial_interpolation_dataset[i].interpolate(method='polynomial', order=degree, inplace=True)
        except ValueError as e:
            pass

    polynomial_interpolation_dataset = polynomial_interpolation_dataset.bfill()
    polynomial_interpolation_dataset = polynomial_interpolation_dataset.ffill()
    all_datasets[f'polynomial_{degree}_interpolation'] = polynomial_interpolation_dataset
```

Rescale data and save datasets

```python
rescaled_datasets = {}
for dataset_name, dataset in all_datasets.items():
    dataset.to_excel(fr'dataset/dataset_versions/{dataset_name}_dataset.xlsx')
    scaler = MinMaxScaler(feature_range=(0, 1))
    rescaled_data = pd.DataFrame(scaler.fit_transform(dataset))
    rescaled_data = rescaled_data.set_axis([column for column in dataset], axis=1)
    rescaled_data = rescaled_data.iloc[1:, :].reset_index().iloc[:, 1:]
    rescaled_data.to_excel(fr'dataset/dataset_versions/{dataset_name}_rescaled_dataset.xlsx')
    rescaled_datasets[dataset_name] = dataset
```

Corr matrix

```python
for dataset_name, dataset in all_datasets.items():
    print(dataset_name)
    corr_matrix = dataset.corr()
    corr_target = corr_matrix["GDP per capita (current US$)"].sort_values(ascending=False)
    print(corr_target)
    corr_target.to_excel(fr'dataset/datasets_corr_target/{dataset_name}_corr_target.xlsx')
    display('')
```
